```python
# applying the label Encoder on the Target value
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
sonar["R"]=encoder.fit_transform(sonar['R'])
```

```python
sonar
```

```python
# using train test split , spliting the data into train data and test data
from sklearn.model_selection import train_test_split
```

```python
x=sonar.drop(columns=['R'])
```

```python
y=sonar['R']
```

```python
x,y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape,y_train.shape
```

```python
print(x_train)
```

```python
print(x_test),x_test.shape
```

```python
# model Training using Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# Checking the accuracy of the model
```

```python
from sklearn.metrics import accuracy_score
```

```python
x_train_prediction=model.predict(x_train)
```

```python
train_data_accuracy = accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on Train data",train_data_accuracy*100)
```

```python
# accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on test data",test_data_accuracy)
```
